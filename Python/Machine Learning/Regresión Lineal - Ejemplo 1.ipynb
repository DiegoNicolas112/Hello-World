### Ejemplo de Regresión Lineal

```python
#Importar las librerías
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Declaramos los vectores de la tabla
# X = Representa la Altura de los individuos.
X = [173, 171, 189, 181]

# Y = Representa los pesos de los inviduos.
y = [81, 72, 96, 94]

# Agregamos un Intercepto o Bias: Este término nos va a permitir desplazar la línea de arriba hacía abajo.
# np.c_ concatena dos vectores y np.ones() crea un vector con la cantidad de unos (1s) que indiquemos (En este caso 4)
X_ones = np.c_[np.ones(4), X]

# Esto nos genera una matriz de 2 dimensiones que contiene los 1s y las alturas ya declaradas.
X_ones


```

```python
# Ahora vamos a sustituir estos vectores en la ecuación normal y para ello utilizaremos numpy.
theta = np.linalg.inv(X_ones.T.dot(X_ones)).dot(X_ones.T).dot(y)
theta
```

```python
# Cálcular el peso de David.
#Primero cálculamos alturas y pesos que ya provienen de la tabla.
plt.scatter(X,y, s=40, c='#06d6a0')


```

```python
# Ahora revisaremos la inclinación de la línea.
X_lim = [170,190]
X_lim_ones = np.c_[np.ones(2), X_lim]
y_lim = X_lim_ones.dot(theta)
plt.plot(X_lim, y_lim, 'r-')

#Configuramos la gráfica
plt.axis([170, 190, 70, 100])
plt.xlabel('Altura')
plt.ylabel('Peso')
plt.title('Altura x Peso')
plt.grid()
```

```python
# Sustituimos los parametros de theta 0 y theta 1 en la fórmula inicial y colocamos la altura de David que ya conocemos
peso_david = theta[0] + (theta[1] * 179)
peso_david
```
