### Regresión Polinomial

```python
# Importamos Librerías
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
m = 100
X = 6 * np.random.rand(m,1) - 3

# y va a ser nuestra función cuadratica.
y = 2 + X + 0.5 * X **2 # 2 + X + 0.5 X 2

plt.scatter(X, y)
plt.show()

```

```python
# Si queremos generar un conjunto de datos mas real realizamos lo siguiente:
y = 2 + X + 0.5 * X **2 + np.random.rand(m,1) # 2 + X + 0.5 X 2

plt.scatter(X, y)
plt.show()
```

```python
# Ahora generaremos atributos polinomiales a través de X
m = 100
X = 6 * np.random.rand(m,1) - 3

# y va a ser nuestra función cuadratica.
y = 2 + X + 0.5 * X **2 + np.random.rand(m,1) # 2 + X + 0.5 X 2

poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)

lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)
# print(X[0], X_poly[0])

print(f"{lin_reg.intercept_[0]} + {lin_reg.coef_[0][0]}x + {lin_reg.coef_[0][1]}x2")

# Una vez que ya esta entrenado podemos ver la curva que estimo el modelo
# Debemos crear un conjunto de datos falso solamente para gráficar la línea.
X_plot = np.linspace(-3,3,m).reshape(m,1)
X_plot_poly = poly_features.transform(X_plot)
y_plot = lin_reg.predict(X_plot_poly)
# print(X_plot_poly)

plt.scatter(X, y)
plt.plot(X_plot, y_plot, 'r')
plt.show()
```
